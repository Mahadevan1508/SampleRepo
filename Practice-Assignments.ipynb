```python
#Temperature Alert System with Conditional Statements
curr_temp = float(input("Enter the current temperature in Celsius: "))
if (curr_temp < 0):
    print("Freezing")
elif (curr_temp > 0 and curr_temp < 10):
    print("Cold")
elif (curr_temp > 10 and curr_temp < 25):
    print("Moderate")
else:
    print("Hot")
```
