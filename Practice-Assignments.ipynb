```python
#Scope-Aware Variable Logger
#nonlocal keyword
def outer():
    x = 10
    def inner():
        nonlocal x
        x = 15
    inner()
    print(x)
outer()
#global keyword
def outer():
    y = 10
    def inner():
        global y
        y = 15
    inner()
    print(y)
outer()
```

```python
#Functions | Data Structures
#Inventory Management System with Lists and List Comprehensions
#restock_book
inventory = [("Python", 10, 29.99),      
             ("Clean Code", 5, 39.99),
             ("Data Science Handbook", 8, 49.99),
             ("Fluent Python", 6, 59.99),
             ("Effective Java", 4, 45.50)
            ]
def restock_book(title,quantity):
    if not isinstance(quantity,int) or quantity <= 0:
        print("Quantity must be a postive integer.")
        return
        
    for book,(t,q,price) in enumerate(inventory):
        if t == title:
            new_quantity = q + quantity
            inventory[book] = (t, new_quantity, price)
            print(f"Restocked {t} book by {quantity}. New quantity: {new_quantity}.")
            return
        
    print(f" {title} book not found in inventory.")
restock_book("Python",1) #Restocks existing book
restock_book("Python",-1)#Handles invalid quantity
restock_book("SQL",1)#Handles book not found

# check updated inventory
print(inventory)
```

```python
#sell_book
inventory = [("Python", 10, 29.99),      
             ("Clean Code", 5, 39.99),
             ("Data Science Handbook", 8, 49.99),
             ("Fluent Python", 6, 59.99),
             ("Effective Java", 4, 45.50)
            ]
def sell_book(title, quantity):
    # Validate input quantity
    if quantity <= 0:
        print("Quantity must be greater than 0.")
        return

    for idx,(t,q,price) in enumerate(inventory):
        if t == title:
            # Check stock
            if q == 0:
                print(f"{t} is out of stock")
                return
            if quantity > q:
                print(f"Only {q} copies are available. Cannot sell {quantity}")
                return
            # Reduce stock correctly
            reduced_quantity = q - quantity
            inventory[idx] = (t, reduced_quantity, price)
            print(f"The {t} book is reduced to {reduced_quantity}")
            return
        # If we finish the loop with no match
    print(f" Sorry {title} Book is not available, Kindly choose the below mentioned  ")
# Example usage:
sell_book("Python Code", 1)
print(inventory)
```

```python
#Apply_Discount
inventory = [("Python", 10, 30),      
             ("Clean Code", 5, 39.99),
             ("Data Science Handbook", 8, 49.99),
             ("Fluent Python", 6, 59.99),
             ("Effective Java", 4, 45.50)
            ]
def apply_discount(category,discount):
    return[
        (title,
         quantity,
         round(price * (1 - discount/100),2)
         if category in title
               else price)
        for title, quantity, price in inventory
    ]
discounted_inventory = apply_discount("Python",10)
print("\n".join(
        f"Title: {title} | Qty: {quantity} | Price: Rs.{price}"
        for title, quantity, price in discounted_inventory
    )
)
```

```python
#Student Grade Analyzer Using Tuples and Dictionaries
student_records =  (
                   (12345,"Madhan", (100,90,88)), #Math, Science, English
                   (67890,"Prajhwal", (95,80,78)),
                   (10293,"Shantanab", (95,30,68)),
                   (99876,"Chaitanya", (75,99,98)),
                   (54321,"Anji", (65,70,58)),
)
                  
def create_student_dict(student_records):
    student_dict = {}
    for student_id, name, score in student_records:
        average_score = sum(score) / len(score)
        student_dict[student_id] = {
            "name": name,
            "average_score": average_score
        }
    return student_dict
    
# dictionary:
# {
#   12345: ('Madhan', (100, 90, 88)),
#   67890: ('Prajhwal', (95, 80, 78)),
#   10293: ('Shantanab', (95, 30, 68)),
#   99876: ('Chaitanya', (75, 99, 98)),
#   54321: ('Anji', (65, 70, 58))
# }

def find_top_student(student_dict):
    max_avg = max(student["average_score"] for student in student_dict.values())

    top_students = []
    for student_id, student in student_dict.items():
        if student["average_score"] == max_avg:
            top_students.append({
                "student_id": student_id,
                "name": student["name"],
                "average_score": student["average_score"]
            })

    return top_students

def get_student_performace(category):
    subject_index = {
        'math': 0,
        'science':1,
        'english':2
    }
    category = category.lower()
    if category not in subject_index:
        print("Category must be 'math', 'science', 'english'")

    index = subject_index[category]

    return {
        name: score[index]
        for _, name, score in student_records
    }

def add_new_student(student_records):
    new_student = (57913,"Prasanna", (60,75,50))
    overall_student = student_records+(new_student,)
    return overall_student
    
student_dict = create_student_dict(student_records)
print("Student Dictionary:")
print(student_dict)
top_students = find_top_student(student_dict)
print("\nTop Student")
print(top_students)
print("\nMath Scores:")
print(get_student_performace('math'))
print("\nScience Scores:")
print(get_student_performace('science'))
print("\nEnglish Scores:")
print(get_student_performace('english'))
print('\nOverall Student:')
overall_student = add_new_student(student_records)
print(overall_student)
```

```python
#Password Strength Checker with Advanced String Operations
# def check_length(password):
#     if len(password) <= 8:
#         print("Password must be eight characters")
#         return
# print(check_length("Shanparmat"))
# print(check_length("asfhj"))
def check_length(password):
    return len(password) >= 8

print("\nCheck Length")
print(check_length("hello"))        # False
print(check_length("password"))     # True
print(check_length("pass1234"))     # True

def check_case(password):
    has_upper = False
    has_lower = False

    for character in password:
        if character.isupper():
            has_upper = True
        elif character.islower():
            has_lower = True
    return has_upper and has_lower

print("\nCheck Case")
print(check_case("hello"))        # False
print(check_case("PASSWORD"))     # False
print(check_case("PASSword"))     # True

def check_digits(password):
    return any(character.isdigit() for character in password)
print("\nCheck Digits")
print(check_digits("pass123"))
print(check_digits("password"))
print(check_digits("pass@orld"))

def check_special_Chars(password):
    special_chars = {'!','@','#','$','%','^','&','*'} #set
    password_chars = set(password)

    return bool(special_chars.intersection(password_chars))
print('\nCheck Special_Character')
print(check_special_Chars('pass@orld'))
print(check_special_Chars('PASS23'))
print(check_special_Chars('p@$$#or!@'))

def password_strength(password):
    checks = {
        "length" : check_length(password),
        "case"  : check_case(password),
        "digits" : check_digits(password),
        "special": check_special_Chars(password)
    }
    passed = sum(checks.values())        

    total = len(checks)

    if passed == total:
        return "strong"
    elif passed >= total -1:
        return "medium"
    else:
        return "weak"
print('\nPassword Strength')
print(password_strength("Shanparmat@1"))
print(password_strength("shanparmat@1"))
print(password_strength("shan"))
```

```python
#Building a Bank Account System with Classes and Objects
#using instance(object) method
class BankAccount:
    def __init__(self,account_number:int,account_holder:str,balance = 0.0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.__balance = balance #private attribute
        #Deposit Method
    def deposit(self,amount:float):
        if amount <= 0:
            print("Deposit amount must be positive.")
            return

        self.__balance += amount
        print(f"Deposit Successful! New balance is {self.__balance:.2f}")
        #Withdrawal Method
    def withdraw(self,withdraw_amount:float):
        if withdraw_amount <= 0:
            print("withdrawal amount must be positive.")
            return
        if withdraw_amount > self.__balance:
            print("Insufficient Balance")
            return
            
        self.__balance -= withdraw_amount
        print(f"Withdraw Successful! New balance is {self.__balance:.2f}")
    def display_balance(self):
        print(
            f"Account Number: {self.account_number}\n"
            f"Account Holder: {self.account_holder}\n"
            f"Current Balance: {self.__balance:.2f}"
        )
account1 = BankAccount(12345,"Madhan",1000)
account2 = BankAccount(67890,"Santhosh",2000)
account3 = BankAccount(10293,"Prajhwal",3000)
account1.deposit(1000)
account1.withdraw(500)
account1.display_balance()
account2.deposit(-1000)
account2.withdraw(500)
account2.display_balance()
account3.deposit(1000)
account3.withdraw(500)
account3.display_balance()
```
