```python
#Function-Driven Calculator
def add(a,b):
    return a+b
def subtract(a,b):
    return a-b
def multiply(a,b):
    return a*b
def divide(a,b):
    if b == 0:
        return("Divisible by zero is not allowed")
    return a/b

    #Dictionary mapping user choices to functions
operations = {
    "1":add,
    "2":subtract,
    "3":multiply,
    "4":divide,
}
print("Simple Calculator")
print("-----------------")

while True:
    print("\nChoose an operation: ")
    print("1.Add")
    print("2.Subtract")
    print("3.Multiply")
    print("4.Divide")
    print("Type 'quit' to exit.")

    choice = input("Enter your choice: ")
    
    if choice.lower() == "quit":
        print("Exiting calculator...Goodbye!")
        break
    if choice not in operations:
        print("Invalid choice! Please try again.")
        continue

    #Getting Operands
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))

    #call the function from dictionary
    result = operations[choice](a,b)
    print("Results: ", result)
```

```python
#Lambda-Based Data Filtering
employees = [
    {"name": "Arun", "salary": 45000, "department": "HR"},
    {"name": "Meena", "salary": 52000, "department": "Finance"},
    {"name": "Ravi", "salary": 48000, "department": "IT"},
    {"name": "Kavya", "salary": 60000, "department": "IT"},
    {"name": "Suresh", "salary": 39000, "department": "Sales"},
    {"name": "Priya", "salary": 55000, "department": "Marketing"},
    {"name": "Vijay", "salary": 47000, "department": "Support"},
    {"name": "Latha", "salary": 58000, "department": "Finance"},
    {"name": "Karthik", "salary": 62000, "department": "IT"},
    {"name": "Nisha", "salary": 43000, "department": "HR"}
]
salary_filter =list(filter(lambda employee : employee["salary"] == 52000, employees))
print(salary_filter)
department_filter =list(filter(lambda employee : employee["department"] == "HR", employees))
print(department_filter)
Name_filter =list(filter(lambda employee : employee["name"].startswith("A"), employees))
print(Name_filter)
min_salary = int(input("Enter Minimum Salary: "))
min_salary_filter = list(filter(lambda employee: employee["salary"] <= min_salary, employees))
print(min_salary_filter)
```

```python
#Scope-Aware Variable Logger
#nonlocal keyword
def outer():
    x = 10
    def inner():
        nonlocal x
        x = 15
    inner()
    print(x)
outer()
#global keyword
def outer():
    y = 10
    def inner():
        global y
        y = 15
    inner()
    print(y)
outer()
```

```python
#Functions | Data Structures
#Inventory Management System with Lists and List Comprehensions
#restock_book
inventory = [("Python", 10, 29.99),      
             ("Clean Code", 5, 39.99),
             ("Data Science Handbook", 8, 49.99),
             ("Fluent Python", 6, 59.99),
             ("Effective Java", 4, 45.50)
            ]
def restock_book(title,quantity):
    if not isinstance(quantity,int) or quantity <= 0:
        print("Quantity must be a postive integer.")
        return
        
    for book,(t,q,price) in enumerate(inventory):
        if t == title:
            new_quantity = q + quantity
            inventory[book] = (t, new_quantity, price)
            print(f"Restocked {t} book by {quantity}. New quantity: {new_quantity}.")
            return
        
    print(f" {title} book not found in inventory.")
restock_book("Python",1) #Restocks existing book
restock_book("Python",-1)#Handles invalid quantity
restock_book("SQL",1)#Handles book not found

# check updated inventory
print(inventory)
```

```python
#sell_book
inventory = [("Python", 10, 29.99),      
             ("Clean Code", 5, 39.99),
             ("Data Science Handbook", 8, 49.99),
             ("Fluent Python", 6, 59.99),
             ("Effective Java", 4, 45.50)
            ]
def sell_book(title, quantity):
    # Validate input quantity
    if quantity <= 0:
        print("Quantity must be greater than 0.")
        return

    for idx,(t,q,price) in enumerate(inventory):
        if t == title:
            # Check stock
            if q == 0:
                print(f"{t} is out of stock")
                return
            if quantity > q:
                print(f"Only {q} copies are available. Cannot sell {quantity}")
                return
            # Reduce stock correctly
            reduced_quantity = q - quantity
            inventory[idx] = (t, reduced_quantity, price)
            print(f"The {t} book is reduced to {reduced_quantity}")
            return
        # If we finish the loop with no match
    print(f" Sorry {title} Book is not available, Kindly choose the below mentioned  ")
# Example usage:
sell_book("Python Code", 1)
print(inventory)
```

```python
#generate_report
inventory = [("Python", 10, 29.99),      
             ("Clean Code", 5, 39.99),
             ("Data Science Handbook", 8, 49.99),
             ("Fluent Python", 6, 59.99),
             ("Effective Java", 4, 45.50)
            ]
def generate_report(inventory):
    return [
        f"{title} - Quantity: {quantity} - Low Stock"
        for title, quantity, price in inventory
        if quantity < 5
    ]
generate_report(inventory)
```

```python
#Apply_Discount
inventory = [("Python", 10, 30),      
             ("Clean Code", 5, 39.99),
             ("Data Science Handbook", 8, 49.99),
             ("Fluent Python", 6, 59.99),
             ("Effective Java", 4, 45.50)
            ]
def apply_discount(category,discount):
    return[
        (title,
         quantity,
         round(price * (1 - discount/100),2)
         if category in title
               else price)
        for title, quantity, price in inventory
    ]
discounted_inventory = apply_discount("Python",10)
print("\n".join(
        f"Title: {title} | Qty: {quantity} | Price: Rs.{price}"
        for title, quantity, price in discounted_inventory
    )
)
```

```python
#Building a Bank Account System with Classes and Objects
#using instance(object) method
class BankAccount:
    def __init__(self,account_number:int,account_holder:str,balance = 0.0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.__balance = balance #private attribute
        #Deposit Method
    def deposit(self,amount:float):
        if amount <= 0:
            print("Deposit amount must be positive.")
            return

        self.__balance += amount
        print(f"Deposit Successful! New balance is {self.__balance:.2f}")
        #Withdrawal Method
    def withdraw(self,withdraw_amount:float):
        if withdraw_amount <= 0:
            print("withdrawal amount must be positive.")
            return
        if withdraw_amount > self.__balance:
            print("Insufficient Balance")
            return
            
        self.__balance -= withdraw_amount
        print(f"Withdraw Successful! New balance is {self.__balance:.2f}")
    def display_balance(self):
        print(
            f"Account Number: {self.account_number}\n"
            f"Account Holder: {self.account_holder}\n"
            f"Current Balance: {self.__balance:.2f}"
        )
account1 = BankAccount(12345,"Madhan",1000)
account2 = BankAccount(67890,"Santhosh",2000)
account3 = BankAccount(10293,"Prajhwal",3000)
account1.deposit(1000)
account1.withdraw(500)
account1.display_balance()
account2.deposit(-1000)
account2.withdraw(500)
account2.display_balance()
account3.deposit(1000)
account3.withdraw(500)
account3.display_balance()
```
